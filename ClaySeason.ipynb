```python
import glob

# Define the year range
years = range(2010, 2025)  # Includes 2000 to 2024

# Create a list to store DataFrames
df_list = []

# Iterate through the years and read CSV files
for year in years:
  file_path = f'/content/drive/My Drive/tennis_data/atp_matches_{year}.csv'  # Use f-strings
  try:
    df = pd.read_csv(file_path)
    df_list.append(df)
  except FileNotFoundError:
    print(f"File not found for year: {year}")  # Handle missing files gracefully

# Concatenate all DataFrames
matches = pd.concat(df_list, ignore_index=True)

# Display the combined DataFrame
matches
```

```python
matches = matches[matches['surface'] == 'Clay']
essential_columns = [
    'tourney_id', 'tourney_name', 'draw_size', 'tourney_level',
    'winner_id', 'loser_id', 'winner_entry', 'loser_entry',
    'winner_ioc', 'loser_ioc', 'winner_rank_points', 'loser_rank_points',
    'match_num'
]
matches.dropna(subset=essential_columns, inplace=True)
matches['tourney_date'] = pd.to_datetime(matches['tourney_date'], format='%Y%m%d')

# winner_name loser_name
matches['winner_name'] = matches['winner_name'].str.split(', ').str[::-1].str.join(' ')
matches['loser_name'] = matches['loser_name'].str.split(', ').str[::-1].str.join(' ')
matches
```
